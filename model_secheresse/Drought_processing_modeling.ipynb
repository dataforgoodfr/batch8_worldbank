### Include SPI indice calculated from spi R package 

Main advantages of the SPI (Standardized Precipitation Index) :

- requires only monthly precipitation;

- can be compared for regions with different climates;

- the normalization of the index makes it possible to determine the rarity of a drought.

The use of this index is also recommended by the World Meteorological Organization (WMO).

Values and meaning of the SPI :

SPI > 0: more precipitation than normal (wetter) ;

SPI < 0: less precipitation than normal (drier) ;

-0.99 < SPI < +0.99: precipitation close to normal ;

SPI < - 2.0: extremely dry ;

SPI > 2.0: extremely wet.

```python
###Financial and human impacts
impacts = (
    df.groupby(["Region"])
    .agg({"Total_Affected": "mean", "Total_Damages": "mean", "Disaster_Type": "count"})
    .reset_index()
)
```

```python
impacts["avg_financial_impacts"] = impacts["Total_Damages"] / impacts["Disaster_Type"]
impacts["avg_human_impacts"] = impacts["Total_Affected"] / impacts["Disaster_Type"]
```

```python
impacts["avg_financial_impacts"] = (
    impacts["avg_financial_impacts"].fillna(0).astype("int64")
)
impacts["avg_human_impacts"] = impacts["avg_human_impacts"].fillna(0).astype("int64")
```

```python
impacts.head()
```

```python
impacts_by_region = impacts[["Region", "avg_financial_impacts", "avg_human_impacts"]]
impacts_by_region = impacts_by_region.rename(
    columns={"Region": "UN_Geosheme_Subregion"}
)
impacts_by_region.head()
```

```python
impacts_by_region_preds = pd.merge(
    final_predictions, impacts_by_region, how="left", on="UN_Geosheme_Subregion"
)
```

```python
impacts_by_region_preds.tail()
```

```python
impacts_by_region_preds["Human_Impact"] = (
    impacts_by_region_preds["DO"] * impacts_by_region_preds["avg_human_impacts"]
)
impacts_by_region_preds["Financial_Impact"] = (
    impacts_by_region_preds["DO"]
    * impacts_by_region_preds["avg_financial_impacts"]
    * 1000
)
```
