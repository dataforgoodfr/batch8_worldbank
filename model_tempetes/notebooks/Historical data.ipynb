# Join to get historical storms

```python
storms_historical = pd.merge(storms, region_iso_clean, how="left", on="ISO")
storms_historical.head()
```

```python
storms_historical["decade"] = storms_historical["Start Year"].apply(lambda row: (row//10)*10)
```

```python
storms_historical[storms_historical["ISO"]=="RUS"].head(30)
```

```python
storms_historical["ISO"] = storms_historical["ISO"].apply(lambda row: "RUS" if (row=="SUN") else row)
storms_historical["Region"] = storms_historical["Region"].apply(lambda row: row if row!="Russian Federation" else "Eastern Europe")
```

# Average temperature

```python
storms_historical_by_region = storms_historical.groupby(["Region", "decade"]).agg({"Disaster Type":"count",
                                                                                   "No Affected":"sum",
                                                                                   "Total Damages ('000 US$)":"sum",
                                                                                   "avg_monthly_temperature_mean":"mean"
                                                                                  }).reset_index()
```

```python
storms_historical_by_region.head()
```

```python
storms_historical_by_region.shape
```

```python
storms_historical_by_region= storms_historical_by_region.rename(columns={"Region":"UN_Geosheme_Subregion",
                                          "decade":"Decade",
                                          "Disaster Type":"DO",
                                          "No Affected":"Human_Impact",
                                          "Total Damages ('000 US$)":"Financial_Impact",
                                          "avg_monthly_temperature_mean":"Â°C"})
storms_historical_by_region.head()
```
