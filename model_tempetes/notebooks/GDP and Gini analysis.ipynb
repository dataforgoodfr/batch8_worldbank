# Import emdat

```python
path = "C:\\batch8_worldbank\\datasets\\tempetes\\wb_disasters_bdd.xlsx"
```

```python
disasters_df = pd.read_excel(path)
disasters_df.head()
```

# Filter on storms

```python
storms_df = disasters_df[disasters_df["Disaster Type"] == "Storm"]
```

```python
nb_storms_by_year_by_country = storms_df.groupby(["Start Year", "ISO"]).aggregate(
    {"Disaster Type": "count", "No Affected": "sum", "Total Damages ('000 US$)": "sum"}
)
nb_storms_by_year_by_country = nb_storms_by_year_by_country.reset_index()
nb_storms_by_year_by_country = nb_storms_by_year_by_country.rename(
    columns={
        "Start Year": "year",
        "Disaster Type": "storms_count",
        "No Affected": "total_nb_affected",
        "Total Damages ('000 US$)": "total_damages",
    }
)
```

```python
nb_storms_by_year_by_country["decade"] = nb_storms_by_year_by_country["year"].apply(
    lambda row: (row // 10) * 10
)
nb_storms_by_decade_by_country = nb_storms_by_year_by_country.groupby(
    ["decade", "ISO"]
).aggregate({"storms_count": "sum", "total_nb_affected": "sum", "total_damages": "sum"})
nb_storms_by_decade_by_country = nb_storms_by_decade_by_country.reset_index()
```

```python
nb_storms_by_decade_by_country.head()
```

```python
nb_storms_by_decade_by_country_after_1960 = nb_storms_by_decade_by_country[
    nb_storms_by_decade_by_country["decade"] >= 1960
]
```

# Join on country code and decade 3 tables

```python
# gdp_pivot, gini_pivot
```

```python
nb_storms_by_decade_by_country_after_1960.dtypes
```

```python
gdp_pivot.dtypes
```

```python
gdp_pivot.shape
```

```python
gdp_pivot["decade"] = gdp_pivot["Date"].apply(lambda row: (row // 10) * 10)
gdp_pivot_by_decade_by_country = gdp_pivot.groupby(
    ["decade", "Country Code"]
).aggregate({"gdp_capita": "mean"})
gdp_pivot_by_decade_by_country = gdp_pivot_by_decade_by_country.reset_index()
```

```python
gdp_pivot_by_decade_by_country.shape
```

```python
gdp_pivot_by_decade_by_country.head()
```

```python
nb_storms_by_decade_by_country_after_1960_gdp = pd.merge(
    nb_storms_by_decade_by_country_after_1960,
    gdp_pivot_by_decade_by_country,
    how="left",
    left_on=["decade", "ISO"],
    right_on=["decade", "Country Code"],
)
nb_storms_by_decade_by_country_after_1960_gdp = (
    nb_storms_by_decade_by_country_after_1960_gdp.drop(["Country Code"], axis=1)
)
```

```python
nb_storms_by_decade_by_country_after_1960_gdp.head()
```

```python
gini_pivot["decade"] = gini_pivot["Date"].apply(lambda row: (row // 10) * 10)
gini_pivot_by_decade_by_country = gini_pivot.groupby(
    ["decade", "Country Code"]
).aggregate({"gini_index": "mean"})
gini_pivot_by_decade_by_country = gini_pivot_by_decade_by_country.reset_index()
```

```python
gini_pivot_by_decade_by_country.head()
```

```python
nb_storms_by_decade_by_country_after_1960_gdp_gini = pd.merge(
    nb_storms_by_decade_by_country_after_1960_gdp,
    gini_pivot_by_decade_by_country,
    how="left",
    left_on=["decade", "ISO"],
    right_on=["decade", "Country Code"],
)
nb_storms_by_decade_by_country_after_1960_gdp_gini = (
    nb_storms_by_decade_by_country_after_1960_gdp_gini.drop(["Country Code"], axis=1)
)
```

```python
nb_storms_by_decade_by_country_after_1960_gdp_gini.head()
```

# Correlations

```python
nb_storms_by_decade_by_country_after_1960_gdp_gini_cor = (
    nb_storms_by_decade_by_country_after_1960_gdp_gini[
        [
            "ISO",
            "storms_count",
            "total_nb_affected",
            "total_damages",
            "gdp_capita",
            "gini_index",
        ]
    ]
)
corr = nb_storms_by_decade_by_country_after_1960_gdp_gini_cor.corr()
sm.graphics.plot_corr(corr, xnames=list(corr.columns))
plt.show()
```
